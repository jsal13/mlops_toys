```python
import whylogs as why
import pandas as pd

import numpy as np
```

```python
num_rows = 10000
feature_1 = np.random.normal(size=num_rows)
error = np.random.normal(scale=1, size=num_rows)
target = 2 * feature_1 + error

df = pd.DataFrame({"feature_1": feature_1, "target": target})
```

```python
df.plot(x="feature_1", y="target", kind="scatter")
```

```python
# Log the results.
# why.init()
results = why.log(df)
```
